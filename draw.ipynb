```python
import numpy as np
import matplotlib.pyplot as plt
x_model_names = ['ResNet50', 'ResNet50+wingloss',
                 'ResNet50+softwingloss', 'hrnetV2',
                 'hrnetV2+awing', 'hrnetV2+dark']
x = range(len(x_model_names))
y1 = [4.86, 4.41, 4.64, 4.02, 4.06, 3.98]
y2 = [5.681, 5.901, 6.061, 6.335, 6.372, 6.424]
y3 = [9.90, 9.25, 8.51, 6.89, 6.80, 6.80]
l1 = plt.plot(x,y1,'ro-', label='NME')
l2 = plt.plot(x, y2, 'g+-', label='0.1 * AUC@0.1')
l3 = plt.plot(x, y3, 'b^-', label='FR@0.1')
plt.plot(x, y1, 'ro-')
plt.plot(x, y2, 'g+-')
plt.plot(x, y3, 'b^-')
plt.xticks(x, x_model_names, rotation=45)
plt.xlabel('model')
plt.ylabel('%')
plt.legend()
plt.show()
```
